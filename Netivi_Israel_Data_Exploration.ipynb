```
# שלב 1: התקנה וייבוא
!pip install -q pandas geopandas matplotlib
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt

print("הספריות נטענו בהצלחה.")

# שלב 2: הגדרת נתיבים
BASE_PATH = '/content/drive/MyDrive/National_model_v0.33_open_data'
RESOURCES_PATH = f'{BASE_PATH}/resources'
SCENARIO_PATH = f'{BASE_PATH}/2020'
NETWORK_PATH = f'{SCENARIO_PATH}/Network'

# שלב 3: בניית טבלאות הנתונים מהקבצים הנכונים
try:
    # --- 3.1: בניית טבלת הצמתים (nodes_df) מ-base_network.in ---
    nodes_data = []
    with open(f'{NETWORK_PATH}/base_network.in', 'r') as f:
        for line in f:
            line = line.strip()
            # שורות הגדרת צומת מתחילות ב-'a '
            if line.startswith('a '):
                parts = line.split()
                # המבנה הוא: a node_id x_coord y_coord ...
                nodes_data.append({
                    'node_id': int(parts[1]),
                    'x_coord': float(parts[2]),
                    'y_coord': float(parts[3])
                })
    nodes_df = pd.DataFrame(nodes_data)
    print(f"טבלת צמתים (nodes_df) נוצרה בהצלחה מ-base_network.in. {len(nodes_df)} צמתים נמצאו.")

    # --- 3.2: טעינת טבלת הכבישים (links_df) מ-calcparam.csv ---
    # קובץ זה מכיל את החיבורים (i, j) ואת הפרמטרים
    links_df = pd.read_csv(f'{NETWORK_PATH}/calcparam.csv')
    links_df.rename(columns={'inode': 'i', 'jnode': 'j'}, inplace=True)

    # --- 3.3: הוספת נתוני נפח ונתיבים חסרים (כהנחות) ---
    # בהיעדר קובץ link_calcparam.csv, ניצור את העמודות החסרות עם הנחות
    if 'lanes' not in links_df.columns:
      links_df['lanes'] = 1 # הנחה: נתיב אחד כברירת מחדל
    if 'volam' not in links_df.columns:
      # נשתמש בעמודת הקיבולת אם קיימת, אחרת נשים ערך קבוע
      if '@LinkCapacity' in links_df.columns:
        links_df['volam'] = links_df['@LinkCapacity'] * 0.5 # הנחה: נפח הוא 50% מהקיבולת
      else:
        links_df['volam'] = 500 # הנחת ברירת מחדל אם אין נתוני קיבולת

    # נשנה את שם עמודת הקיבולת לשם אחיד
    if '@LinkCapacity' in links_df.columns:
        links_df.rename(columns={'@LinkCapacity': 'capacity'}, inplace=True)
    else:
        links_df['capacity'] = 1000 # הנחת ברירת מחדל

    print(f"טבלת כבישים (links_df) נטענה בהצלחה. {len(links_df)} קטעי כביש נמצאו.")

    # טעינת אזורי התנועה לויזואליזציה בלבד
    taz_gdf = gpd.read_file(f'{RESOURCES_PATH}/geo/NewTazArzi_848.shp')

except Exception as e:
    print(f"שגיאה קריטית בשלב בניית הנתונים: {e}")

# שלב 4: ניתוח וויזואליזציה
# =========================================================
TARGET_NODE_ID = 100003 # נבחר צומת מהפלט הקודם
TIME_PERIOD = 'AM'
# =========================================================

if 'links_df' in locals() and 'nodes_df' in locals():
    incoming_links_df = links_df[links_df['j'] == TARGET_NODE_ID].copy()
    volume_column = f'vol{TIME_PERIOD.lower()}'

    if not incoming_links_df.empty:
        simulation_params = incoming_links_df[[
            'i', 'j', 'lanes', 'capacity', volume_column
        ]].rename(columns={
            'i': 'צומת_מוצא', 'j': 'צומת_יעד_(המטרה)', 'lanes': 'מספר_נתיבים_(תורים)',
            'capacity': 'קיבולת_שעתית_(בסיס_ל-μ)', volume_column: f'נפח_שעתי_משוער'
        })

        print(f"\n--- פרמטרים שחולצו עבור צומת מספר {TARGET_NODE_ID} ---")
        display(simulation_params)

        target_node_coords = nodes_df[nodes_df['node_id'] == TARGET_NODE_ID]

        if not target_node_coords.empty:
            fig, ax = plt.subplots(1, 1, figsize=(12, 12))
            taz_gdf.plot(ax=ax, color='lightgrey', edgecolor='white', zorder=1)

            ax.scatter(target_node_coords['x_coord'], target_node_coords['y_coord'],
                       color='red', s=100, zorder=5, label=f'צומת מטרה ({TARGET_NODE_ID})')

            for index, link in simulation_params.iterrows():
                origin_node_coords = nodes_df[nodes_df['node_id'] == link['צומת_מוצא']]
                if not origin_node_coords.empty:
                    start_x, start_y = origin_node_coords.iloc[0]['x_coord'], origin_node_coords.iloc[0]['y_coord']
                    end_x, end_y = target_node_coords.iloc[0]['x_coord'], target_node_coords.iloc[0]['y_coord']
                    ax.plot([start_x, end_y], [start_y, end_y], 'b-', linewidth=2, zorder=3)

            plt.title(f"ויזואליזציה של כבישים נכנסים לצומת {TARGET_NODE_ID}")
            ax.legend([plt.Line2D([0], [0], color='r', marker='o', linestyle=''), plt.Line2D([0], [0], color='b', lw=2)],
                      [f'צומת מטרה ({TARGET_NODE_ID})', 'כבישים נכנסים'])

            buffer = 0.1 # הגדרת טווח במעלות (קואורדינטות גיאוגרפיות)
            ax.set_xlim(target_node_coords.iloc[0]['x_coord'] - buffer, target_node_coords.iloc[0]['x_coord'] + buffer)
            ax.set_ylim(target_node_coords.iloc[0]['y_coord'] - buffer, target_node_coords.iloc[0]['y_coord'] + buffer)
            plt.xlabel("קואורדינטת X (Longitude)")
            plt.ylabel("קואורדינטת Y (Latitude)")
            plt.grid(True)
            plt.show()
    else:
        print(f"לא נמצאו כבישים נכנסים לצומת {TARGET_NODE_ID}. נסה מספר צומת אחר.")
else:
    print("הניתוח לא יכול להמשיך מכיוון שאחד או יותר מקבצי הנתונים לא נוצרו בהצלחה.")
```
